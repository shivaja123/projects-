# 20.classify movies based on rating

```python
def rating(rating):
    if rating>=7.0:
        return "Excellent"
    elif rating>6.0:
        return "Good"
    else:
        return "Average"
```

```python
data['Rating_cat']=data['Rating'].apply(rating)
data.head()
```

```python
sns.violinplot(x='Rating_cat',y='Rating',data=data)
plt.title('rating catigores')
plt.show()
```

# 22.Find unique values from genre

```python
data['Genre']
```

```python
list1=[]
for value in data['Genre']:
    list1.append(value.split(','))
list1    
```

```python
one_d=[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
uni_list=[]
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
```
